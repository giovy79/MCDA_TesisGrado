##Análisis de datos faltantes

Los datos faltantes son las respuestas con valores negativos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definir una función para calcular la proporción de valores negativos
def calcular_proporciones_negativas(df):
    df = df.copy()
    df[df < 0] = pd.NA
    proporciones = df.isna().mean() * 100
    return proporciones

def filtrar_mayores_a_10(proporciones):
    return proporciones[proporciones > 10]

# Lista de DataFrames
dfs = [
    dfSocialValues, dfWellbeing, dfsocialcapital,
    dfeconomicvalues, dfcorruption, dfmigration,
    dfsecurity, dfpostmaterialist, dfsciencetechnology,
    dfreligiousvalues, dfethicalvalues, dfpoliticalinterest,
    dfpoliticalculture, dfdemography
]

# Lista de nombres para cada DataFrame
nombres = [
    'Social Values', 'Wellbeing', 'Social Capital',
    'Economic Values', 'Corruption', 'Migration',
    'Security', 'Post Materialist', 'Science Technology',
    'Religious Values', 'Ethical Values Norms', 'Political Interest Participation',
    'Political Culture Regimes', 'Demography'
]

# Crear gráficos separados para cada categoría
for df, nombre in zip(dfs, nombres):
    proporciones = calcular_proporciones_negativas(df)
    proporciones = proporciones.sort_values(ascending=False)

    plt.figure(figsize=(15, 5))
    sns.barplot(x=proporciones.index, y=proporciones.values, hue=proporciones.index, dodge=False, palette='viridis')
    plt.legend([], [], frameon=False)  # Remover la leyenda ya que no es necesaria
    plt.xticks(rotation=45)
    plt.title(f'Proporción de valores negativos por pregunta en la categoría {nombre}')
    plt.xlabel('Pregunta')
    plt.ylabel('Proporción de valores negativos (%)')
    plt.show()


# Crear un diccionario para almacenar las tablas de frecuencia filtradas
tablas_frecuencia = {}

# Calcular y filtrar proporciones para cada DataFrame
for df, nombre in zip(dfs, nombres):
    proporciones = calcular_proporciones_negativas(df)
    proporciones_filtradas = filtrar_mayores_a_10(proporciones)
    tablas_frecuencia[nombre] = proporciones_filtradas

# Mostrar las tablas de frecuencia filtradas
for nombre, tabla in tablas_frecuencia.items():
    print(f"Categoría: {nombre}")
    print(tabla)
    print("\n")
```
