##Analisis de composición de clusteres

```python
import pandas as pd
import umap
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# Crear DataFrame con todas las preguntas del estudio, imputadas, sin estandarizar
data = dfWVSPreguntasEstudio_imputed

df = pd.DataFrame(data)

# Aplicar UMAP
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
embedding_SinEst = reducer.fit_transform(df)

# Convertir a DataFrame para facilitar el manejo
embedding_df_SinEst = pd.DataFrame(embedding_SinEst, columns=['UMAP1', 'UMAP2'])

# Aplicar K-means con 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters_SinEst = kmeans.fit_predict(embedding_SinEst)
embedding_df_SinEst['Cluster'] = clusters_SinEst

# Añadir los clusters al DataFrame original
df['Cluster'] = clusters_SinEst

# Calcular la prevalencia por cluster para las preguntas originales
prevalence_by_cluster = df.groupby('Cluster').mean()

# Determinar el cluster con la mayor prevalencia para cada pregunta
max_prevalence_cluster = prevalence_by_cluster.idxmax(axis=0)

# Contar la cantidad de preguntas con mayor prevalencia en cada cluster
cluster_counts = max_prevalence_cluster.value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
cluster_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Clusters')
plt.ylabel('Número de Preguntas')
plt.title('Número de Preguntas con Mayor Prevalencia en Cada Cluster')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Determinar el cluster con la mayor prevalencia para cada pregunta
max_prevalence_cluster = prevalence_by_cluster.idxmax(axis=0)

# Crear un DataFrame que contenga las preguntas y su cluster de mayor prevalencia
questions_clusters = pd.DataFrame({'Pregunta': max_prevalence_cluster.index, 'Cluster': max_prevalence_cluster.values})

# Agrupar las preguntas por cluster
questions_per_cluster = questions_clusters.groupby('Cluster')['Pregunta'].apply(list)

# Mostrar las preguntas agrupadas por cluster
for cluster, questions in questions_per_cluster.items():
    print(f"Cluster {cluster}:")
    for question in questions:
        print(f"  - {question}")
    print()

# Crear gráficos de barras horizontales separados por cluster
num_clusters = len(questions_per_cluster)
fig, axes = plt.subplots(nrows=num_clusters, ncols=1, figsize=(10, num_clusters * 5))

for i, (cluster, preguntas) in enumerate(questions_per_cluster.items()):
    axes[i].barh(preguntas, prevalence_by_cluster.loc[cluster, preguntas])
    axes[i].set_title(f'Cluster {cluster}')
    axes[i].set_xlabel('Prevalencia')
    axes[i].set_ylabel('Preguntas')
    axes[i].invert_yaxis()  # Para que las preguntas aparezcan en orden descendente

plt.tight_layout()
plt.show()

```
